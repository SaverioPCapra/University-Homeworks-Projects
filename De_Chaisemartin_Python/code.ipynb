```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
obs = 1001

x = np.linspace(start = -0.5, stop = 0.5, num = obs)

power_values = []

for i in x: 
    power_values.append(stats.norm.cdf(-1.96+i*np.sqrt(500),0, 1)+stats.norm.cdf(-1.96-i*np.sqrt(500),0, 1))
```

```python
data = pd.DataFrame({"Mean":x, "Test": power_values})
```

```python
data.plot(x= "Mean", y = "Test")
```
